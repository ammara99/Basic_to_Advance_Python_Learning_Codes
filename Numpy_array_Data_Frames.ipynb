# Reading data from dictionary and changing the orientation

```python
d={0:[1,2,3], 1:[5,6,8]}

print(d)
```

```python
df=pd.DataFrame(d, index=['0','1','2'])
df
```

```python
df=pd.DataFrame.from_dict(d, orient='index', columns=['A','B','C'])
df
```

```python
x=pd.DataFrame(df)
x
```

```python
x.drop("A", inplace=True, axis=1)
x.drop(x.index[0], inplace=True)
x
```

```python
d={}

n=int(input("Enter range : "))
for i in range(n):
    y=[]
    name=input("Enter name : ")
    age=input("Enter Age : ")
    
    for j in range(2):
        skills=input("Enter your subjetcs: ")
        y.append(skills)
    d[i]=name,age,y

print(d)
```

```python
d={0:["Somia",18,["Data Science", "C++"]], 1:["Alina",18,["Java", "C++"]]}

df=pd.DataFrame(d)
df
```

```python
x=pd.DataFrame(df)
x
```

```python
x.drop("Age",inplace=True, axis=1)
x.drop(x.index[1],inplace=True)

x
```
